## Penanganan Anomali

Meninjau ulang dari temuan di atas bahwa :
- Nilai terkecil dari kolom `price` adalah 0 yang mana perlu ditelusuri lebih lanjut
- Nilai terkecil dari kolom `grand_total` dan `discount_amount` bernilai negatif yang mana hal ini perlu ditelusuri lebih 

```python
#price 0
df[df['price']==0]
```

Terdapat 2215 barang dengan harga 0, hal ini dimungkinkan karena barang tersebut adalah barang promosi sehingga dijual dengan harga 0. Data kondisi ini dibiarkan saja dengan pengisian kolom MV sebesar 0 juga.

```python
idx=df.index[df['price']==0].tolist()
df.loc[idx,'MV']=0
```

```python
#grand total negatif
df[df['grand_total']<0]
```

Kolom grand_total yang bernilai negatif mungkin terjadi karena kesalahan input atau sistem yang ada di layanan _e commerce_. Perhitungan grand_total adalah harga barang dikalikan dengan kuantitas pembelian dikurangi dengan diskon yang diberikan.

```python
idx=df.index[df['grand_total']<0].to_list()
df.loc[idx,'grand_total']=df['price']*df['qty_ordered']-df['discount_amount']
```

Masih terdapat beberapa data yang memiliki kolom grand total negatif hal ini dimungkinkan ketika total harga yang harus dibayarkan kurang dari diskon yang diberikan. Kondisi seperti ini dibiarkan saja.

# Analisa Data

## Trend jumlah pembelian dengan metode COD tiap bulan

```python
df['monthyear'] = pd.to_datetime(df['created_at']).dt.to_period('M')
```

```python
df_cod=df[df['payment_method']=='cod']
df_noncod=df[df['payment_method']!='cod']
```

```python
by_month_cod = pd.to_datetime(df_cod['created_at']).dt.to_period('M').value_counts().sort_index()
by_month_cod.index = pd.PeriodIndex(by_month_cod.index)
df_month_cod = by_month_cod.rename_axis('month').reset_index(name='counts cod')
```

```python
by_month_noncod = pd.to_datetime(df_noncod['created_at']).dt.to_period('M').value_counts().sort_index()
by_month_noncod.index = pd.PeriodIndex(by_month_noncod.index)
df_month_noncod = by_month_noncod.rename_axis('month').reset_index(name='counts noncod')
```

```python
df_line = df_month_cod.copy()
df_line['counts noncod']=df_month_noncod['counts noncod']
```

```python
plt.figure(figsize=(16, 8), dpi=150)
  
plt.plot(df_line.index,df_line['counts cod'],color='green',label='COD')
plt.plot(df_line.index,df_line['counts noncod'],color='red',label='Non-COD')

plt.title('Jumlah Pembelian Berdasarkan Metode Pembayaran')
  
plt.xlabel('Bulan')
plt.xticks(df_line.index, df_line['month'],rotation=45)

plt.legend()
```

Dari grafik di atas terlihat bahwa secara umum jumlah pembelian dengan metode COD lebih sedikit dibandingkan dengan metode selain COD. Terutama pada kondisi ketika dalam suatu bulan sedang terjadi banyak pembelian. Namun terlihat juga bahwa metode COD lebih stabil fluktuasinya dari waktu ke waktu dibandingkan dengan metode selain COD. Secara statistik akan dibuktikan apakah terdapat perbedaan yang signifikan dari jumlah pembelian berdasarkan metode pembayaran.

```python

from scipy.stats import shapiro
display(shapiro(df_line['counts cod']),
shapiro(df_line['counts noncod']))
```

P-value untuk data yang ada tidak berdistribusi normal karena p-value kurang dari 0,05

```python
from scipy.stats import wilcoxon
pvalue=wilcoxon(df_line['counts cod'],df_line['counts noncod'])[1]
alpha=0.05
print('====================')
print(f'''
    
    H0\t: Jumlah pembelian dengan COD sama dengan jumlah pembelian dengan selain COD
    H1\t: Jumlah pembelian dengan tidak sama jumlah pembelian dengan selain COD
    
    alpha\t: {alpha}

    pvalue\t:{pvalue}

    ''')
if pvalue < alpha :
    print(' Kesimpulan : Jumlah pembelian dengan tidak sama jumlah pembelian dengan selain COD')
    print('====================')
else:
    print(' Kesimpulan : Jumlah pembelian dengan COD sama dengan jumlah pembelian dengan selain COD')
    print('====================')
```

Dari grafik di atas terlihat bahwa secara umum jumlah pelanggan yang melakukan pembelian dengan COD tiap bulannya lebih banyak dibandingkan dengan selain COD. P-value untuk data yang ada tidak berdistribusi normal karena p-value kurang dari 0,05

```python
from scipy.stats import wilcoxon
pvalue=wilcoxon(df_line_1['Customer ID'],df_line_2['Customer ID'],alternative='greater')[1]
alpha=0.05
print('====================')
print(f'''
    
    H0\t: Jumlah pelanggan yang menggunakan metode COD tidak lebih banyak dibandingkan jumlah pelanggan metode selain COD
    H1\t: Jumlah pelanggan yang menggunakan metode COD lebih banyak dibandingkan jumlah pelanggan metode selain COD
    
    alpha\t: {alpha}

    pvalue\t:{pvalue}

    ''')
if pvalue < alpha :
    print(' Kesimpulan : Jumlah pelanggan yang menggunakan metode COD lebih banyak dibandingkan jumlah pelanggan metode selain COD')
    print('====================')
else:
    print(' Kesimpulan : Jumlah pelanggan yang menggunakan metode COD tidak lebih banyak dibandingkan jumlah pelanggan metode selain COD')
    print('====================')
```
